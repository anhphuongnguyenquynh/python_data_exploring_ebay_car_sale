# Cleaning column names
Use the DataFrame.columns to change some name of columns:
- yearOfRegistration to registration_year
- monthOfRegistration to registration_month
- notRepairedDamage to unrepaired_damage
- dateCreated to ad_created
The rest of the columnn names from camelCase to snake_case.

```python
autos.columns
```

```python
autos_test = autos.copy()
autos_test_column = autos_test.columns
```

```python
#convert camelCase to snake_case
def change_case(text):
    res = [text[0].lower()]
    for c in text[1:]:
        if c in ('ABCDEFGHIJKLMNOPQRSTUVWXYZ'):
            res.append('_')
            res.append(c.lower())
        else:
            res.append(c)
    return ''.join(res)
```

```python
new_columns = []
for name in autos_test.columns:
    new_name = change_case(name)
    new_columns.append(new_name)
new_columns[7] = "registration_year"
new_columns[9] = "powerPS"
new_columns[12] = "registration_month"
new_columns[14] = "unrepaired_damage"
new_columns[15] = "ad_created"
autos.columns = new_columns
```

## Exploring price by brand

I want to find out the average price of top brand car in this dataset and storing aggregate in dataframe

```python
# Lọc ra top 20 brand xe bán chạy

# Tính giá trung bình của brand xe có giá trị cao
# append vào bảng dataframe
```

```python
# Loc ra top 20 brand xe ban chay
top_20 = filtered_autos['unrepaired_damage'].value_counts(dropna=False).head(20)
top_20
```

```python
#Tinh gia trung binh cua nhung brand co gia tri cao
top_20.index
```

```python
top20 = list(top_20.index)
```

```python
avg_price_name = {}
for name in top20:
    mask = filtered_autos.unrepaired_damage == name
    filtered_name_car = filtered_autos[mask]
    avg_price = filtered_name_car['price'].mean()
    avg_price_name[name]=avg_price
```
